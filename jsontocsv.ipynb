```python
final = pd.DataFrame()

for j in range(1,11):
    rs = requests.get(
        'https://api.themoviedb.org/3/movie/popular?api_key=703f0b4742b9931be466f7027ee52797&page={}'.format(j)
    ).json()

    original_title = []
    release_date = []
    movie_id = []         
    overview = []
    popularity = []
    original_language = []

    for i in rs['results']:
        original_title.append(i['original_title'])
        release_date.append(i['release_date'])
        movie_id.append(i['id'])
        overview.append(i['overview'])
        popularity.append(i['popularity'])
        original_language.append(i['original_language'])

    d = {
        'original_title': original_title,
        'release_date': release_date,
        'id': movie_id,
        'overview': overview,
        'popularity': popularity,
        'original_language': original_language
    }

    df = pd.DataFrame(d)

    # ✅ MUST assign back
    # final = final.append(df, ignore_index=True)

    final = pd.concat([final, df], ignore_index=True)

    
```
